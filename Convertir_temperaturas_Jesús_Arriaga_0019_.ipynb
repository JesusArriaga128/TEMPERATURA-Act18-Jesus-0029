### 1. Fórmulas de conversión

* **Celsius a Fahrenheit:**
    $$°F = (°C \times 1.8) + 32$$

* **Fahrenheit a Celsius:**
    $$°C = \frac{°F - 32}{1.8}$$

---

### 2. Ejemplos Resueltos

**Ejemplo 1: Convertir 100°C (Punto de ebullición del agua) a °F**
1.  **Fórmula:** $°F = (Celsius \times 1.8) + 32$
2.  **Sustitución:** $°F = (100 \times 1.8) + 32$
3.  **Operación 1 (Multiplicación):** $100 \times 1.8 = 180$
4.  **Operación 2 (Suma):** $180 + 32 = 212$
5.  **Resultado:** **212°F**

**Ejemplo 2: Convertir 50°F a °C**
1.  **Fórmula:** $°C = (Fahrenheit - 32) / 1.8$
2.  **Sustitución:** $°C = (50 - 32) / 1.8$
3.  **Operación 1 (Resta):** $50 - 32 = 18$
4.  **Operación 2 (División):** $18 / 1.8 = 10$
5.  **Resultado:** **10°C**
